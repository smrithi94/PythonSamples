```python
import pandas as pd
import folium 
```

```python
df=pd.read_csv('Volcanoes_USA.txt')
```

```python
df
```

```python
lat=list(df['LAT'])
lon=list(df['LON'])
name=list(df['NAME'])
elev=list(df['ELEV'])
def get_colour(elevation):
    if elevation>0 and elevation<=1000:
        return 'green'
    elif elevation>1000 and elevation<=3000:
        return 'red'
    elif elevation>3000:
        return 'blue'
map= folium.Map(location=[38.883333, -77.016667],zoom_start=5,tiles='Mapbox Bright')
for lt,ln,ele,nam in zip(lat,lon,elev,name):
    map.add_child(folium.Marker(location=[lt,ln],popup=folium.Popup(str(ele)),icon=folium.Icon(color=get_colour(ele))))
map
```

```python
lat=list(df['LAT'])
lon=list(df['LON'])
name=list(df['NAME'])
elev=list(df['ELEV'])
def get_colour(elevation):
    if elevation>0 and elevation<=1000:
        return 'green'
    elif elevation>1000 and elevation<=3000:
        return 'red'
    elif elevation>3000:
        return 'blue'
map= folium.Map(location=[38.883333, -77.016667],zoom_start=5,tiles='Mapbox Bright')
for lt,ln,ele in zip(lat,lon,elev):
    map.add_child(folium.CircleMarker(location=[lt,ln],radius=6,popup=str(ele),fill_color=get_colour(ele),color='black',fill_opacity=0.7,fill=True))
map
```
